```python
book = pd.concat([pd.read_csv('got/book'+str(b)+'.csv') for b in range (1,6)] ).reset_index(drop=True)
```

```python
book
```

```python
character = list(book.Source.unique()) + list(book.Target.unique())
character
```

```python
character = list(dict.fromkeys(character))
len(character)
```

```python
bookx = book[['Source','Target']].drop_duplicates().index
bookr = book.iloc[bookx,:].reset_index(drop=True)
```

```python
bookr
```

```python
def find_character(name):
    return character.index(name)
    
```

```python
bookr['Target_point'] = bookr.apply(lambda row: find_character(row.Target), axis=1)
```

```python
bookr['Source_point'] = bookr.apply(lambda row: find_character(row.Source), axis=1)
```

```python
bookr
```

```python
bookr.book.value_counts()
```

```python
character 
```

```python
char_group = []
for c in character:
    index = bookr[bookr.Source == c].index
    if len(index)==0:
        index = bookr[bookr.Target == c].index[0]
    else:
        index = index[0]
    group = bookr.book.loc[index]
    char_group.append([c,group])

char_group
```

```python
bookr.isna().sum()
```

```python
char_group_dict = []
for c in filter(lambda ls: ls[1] == float(1),char_group):
    char_group_dict.append( {"name":c[0] , "group":int(c[1])} )


links = []
for index, row in bookr[bookr.book==1].iterrows():
    links.append( {"source":row.Source_point, "target":row.Target_point, "value":row.weight ,"group":int(row.book)} )

    

```

```python
json_obj = {"nodes":char_group_dict, "links":links}
json_obj
```

```python
import json
with open('GoTbook1.json', 'w') as json_file:
    json.dump(json_obj, json_file)
```
